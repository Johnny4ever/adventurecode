```python
def excelcombine(filepath,shtname,skr=0):
    os.chdir(filepath)
    i=0
    lst=[]
    for each in os.listdir(filepath):
        i+=1
        counter="{}{}".format('counter',i)
        print(filepath+'\\'+each)
        counter=pd.read_excel(filepath+'\\'+each,sheetname=shtname,skiprows=skr)        
        lst.append(counter)
        print("{}{}".format(each, " has been processed!"))
    all=pd.concat(lst)
    return all     
```

```python
def csvcombine(filepath,skr=0):
    os.chdir(filepath)
    i=0
    lst=[]
    for each in os.listdir(filepath):
        i+=1
        counter="{}{}".format('counter',i)
        print(filepath+'\\'+each)
        counter=pd.read_csv(filepath+'\\'+each,skiprows=skr)        
        lst.append(counter)
        print("{}{}".format(each, " has been processed!"))
    all=pd.concat(lst)
    return all   
```
